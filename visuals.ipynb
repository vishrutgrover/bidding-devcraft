```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# Suppress warnings for cleaner output
warnings.filterwarnings('ignore')

# Optional: Set a style for plots
sns.set(style="whitegrid")

file_paths = [
    "/Users/vishrutgrover/coding/Adobe Devcraft PS/finaldata/processed_data/processed_data/df_06.csv",
    "/Users/vishrutgrover/coding/Adobe Devcraft PS/finaldata/processed_data/processed_data/df_07.csv",
    "/Users/vishrutgrover/coding/Adobe Devcraft PS/finaldata/processed_data/processed_data/df_08.csv",
    "/Users/vishrutgrover/coding/Adobe Devcraft PS/finaldata/processed_data/processed_data/df_09.csv",
    "/Users/vishrutgrover/coding/Adobe Devcraft PS/finaldata/processed_data/processed_data/df_10.csv",
    "/Users/vishrutgrover/coding/Adobe Devcraft PS/finaldata/processed_data/processed_data/df_11.csv",
    "/Users/vishrutgrover/coding/Adobe Devcraft PS/finaldata/processed_data/processed_data/df_12.csv"
]

dfs = [pd.read_csv(path) for path in file_paths]
df_combined = pd.concat(dfs, ignore_index=True)
```

```python
df_combined
```

```python
print("\nDataset shape:", df_combined.shape)

print("\nDataFrame Info:")
df_combined.info()

print("\nSummary statistics for numerical columns:")
print(df_combined.describe())
```

```python
missing_counts = df_combined.isnull().sum()
print("\nMissing values per column:")
print(missing_counts[missing_counts > 0])
```

```python
# Identify numerical and categorical columns
num_cols = df_combined.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = df_combined.select_dtypes(include=['object', 'category']).columns.tolist()

print("\nNumerical Columns:", num_cols)
print("Categorical Columns:", cat_cols)

# For array columns, if your arrays are stored as lists or numpy arrays within object columns,
# we can detect them by checking the type of the first non-null element.
array_cols = []
for col in df_combined.columns:
    # Get a sample non-null value from the column
    sample_val = df_combined[col].dropna().iloc[0] if not df_combined[col].dropna().empty else None
    if sample_val is not None and isinstance(sample_val, (list, np.ndarray)):
        array_cols.append(col)

print("Array Columns:", array_cols)
```

```python
for col in cat_cols:
    print(f"\nValue counts for categorical column: {col}")
    print(df_combined[col].value_counts().head(10))
    
    if df_combined[col].nunique() < 20:
        plt.figure(figsize=(8, 4))
        sns.countplot(data=df_combined, x=col, order=df_combined[col].value_counts().index)
        plt.title(f"Distribution of {col}")
        plt.xticks(rotation=45)
        plt.show()
```

```python
for col in num_cols:
    plt.figure(figsize=(10, 4))
    sns.histplot(df_combined[col].dropna(), kde=True, bins=30)
    plt.title(f"Distribution of {col}")
    plt.show()
    
    plt.figure(figsize=(10, 2))
    sns.boxplot(x=df_combined[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
plt.figure(figsize=(12, 10))
corr_matrix = df_combined[num_cols].corr()
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.title("Correlation Matrix of Numerical Columns")
plt.show()
```

```python
# Not really useful I believe, kuch sense bne toh dekho
for col in array_cols:
    # Create a new column with the length of the array (if applicable)
    len_col = col + "_len"
    df_combined[len_col] = df_combined[col].apply(lambda x: len(x) if isinstance(x, (list, np.ndarray)) else np.nan)
    
    # Plot the distribution of array lengths
    plt.figure(figsize=(10, 4))
    sns.histplot(df_combined[len_col].dropna(), bins=30)
    plt.title(f"Length Distribution of Array Column: {col}")
    plt.xlabel("Length")
    plt.show()
```

```python
duplicate_count = df_combined.duplicated().sum()
print("\nNumber of duplicate rows:", duplicate_count)
```

```python
from scipy import stats

# Calculate Z-scores and count outliers per column (using a threshold, e.g., |z| > 3)
for col in num_cols:
    z_scores = np.abs(stats.zscore(df_combined[col].dropna()))
    outliers = np.sum(z_scores > 3)
    print(f"{col}: {outliers} outliers")
```
